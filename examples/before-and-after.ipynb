```python
%matplotlib inline
from pointgrid import align_points_to_grid
from sklearn import datasets
import pandas as pd

# create some fake data and visualize the initial point positions
arr, labels = datasets.make_blobs(n_samples=1000, centers=5)
pd.DataFrame(arr, columns=['x', 'y']).plot.scatter('x', 'y', s=0.5)
```

```python
# update the point positions and show the new point positions
updated = align_points_to_grid(arr)
pd.DataFrame(updated, columns=['x', 'y']).plot.scatter('x', 'y', s=0.5)
```
